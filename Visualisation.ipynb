```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [40,5,60]

plt.plot(x,y)
plt.show()
```

```python
housing = pd.DataFrame({'rooms':[1, 2,2,2,1,3,2],
                        'price':[100,200,300,400,500,600,700]})
```

```python
plt.scatter(housing['rooms'], housing['price'])
plt.show()
```

```python
plt.scatter(housing['rooms'], housing['price'], color='red')
plt.xlim(0,4)
plt.ylim(0,800)

plt.title('rooms vs price')
plt.xlabel('rooms')
plt.ylabel('price')
plt.show()
```
